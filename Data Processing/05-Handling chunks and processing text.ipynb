```python
from transformers import T5ForConditionalGeneration, AutoTokenizer
```

```python
model = T5ForConditionalGeneration.from_pretrained("t5-base")
tokenizer = AutoTokenizer.from_pretrained("t5-base")
```

```python
file_path='/content/drive/MyDrive/Colab Notebooks/Testing Model (Tokanization issue)/filterd_file.txt'
with open(file_path, 'r', encoding='utf-8') as file:
  # Read the contents of the file and concatenate
  file_contents = file.read()
```

```python
def split_text_into_chunks(text, min_chunk_words):
    chunks = []          # Initialize an empty list to store the chunks
    current_chunk = []    # Initialize an empty list to build the current chunk

    # Iterate through each word in the input text
    for word in text.split():
        current_chunk.append(word)  # Add the word to the current chunk
        # Check if the current chunk has at least the specified minimum number of words
        if len(current_chunk) >= min_chunk_words:
            chunks.append(" ".join(current_chunk))  # If the condition is met, add the current chunk to the list
            current_chunk = []  # Start a new chunk with an empty list

    # After processing all words, check if there's any remaining words in the current chunk
    if current_chunk:
        chunks.append(" ".join(current_chunk))  # Add the remaining chunk to the list

    return chunks  # Return the list of chunks
```

```python
chunks=split_text_into_chunks(file_contents,450)
```

```python
def encode_and_summarize_chunks(chunks, tokenizer, model):
    summaries = []
    for chunk in chunks:
        input_ids = tokenizer.encode("summarize: " + chunk, return_tensors="pt", truncation=False)
        summary_ids = model.generate(input_ids, max_length=250, min_length=50 , num_beams=4, early_stopping=True)
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
        summaries.append(summary)
    return summaries
```

```python
final_summary=encode_and_summarize_chunks(chunks,tokenizer,model)
```

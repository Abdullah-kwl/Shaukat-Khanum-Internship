```python
import re

def cleanTxt(text):
    text = re.sub('…', '', text)
    text = re.sub('&amp;', '', text)
    text = re.sub('@[A-Za-z0-9]+', '', text)  # Removing @mentions
    text = re.sub('#', '', text)  # Removing '#' hash tag
    text = re.sub('|', '', text)  # Removing '|' sign
    text = re.sub('-', '', text)  # Removing '-' sign
    text = re.sub('_', '', text)  # Removing '_' sign
    text = re.sub('https?:\/\/\S+', '', text)  # Removing hyperlink
    text = re.sub(' +', ' ', text)  # Removing double spaces
    text = re.sub(' {2,}', ' ', text)  # Removing two or more spaces and replacing with a single space

    return text


def filtering_text(input_file_path, output_file_path):
    # Read the lines from the file and remove duplicates
    with open(input_file_path, 'r') as file:
        lines = file.readlines()
        unique_lines = []

        for line in lines:
            if line not in unique_lines:
                unique_lines.append(line)

    # Clean the text
    cleaned_lines = [cleanTxt(line) for line in unique_lines]

    # Write the cleaned lines back to the file
    with open(output_file_path, 'w') as file:
        file.writelines(cleaned_lines)



# Example usage:
input_file_path = "output_result.txt"  # Replace with the actual input file path
output_file_path = "filterd_file.txt"  # Replace with the desired output file path

filtering_text(input_file_path, output_file_path)
```
